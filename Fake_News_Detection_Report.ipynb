## Abstract

The proliferation of fake news and misinformation across digital platforms has emerged as one of the most pressing challenges of the modern information age. With billions of users actively consuming and sharing news content through social media, online news portals, and messaging applications, the speed at which false information can spread globally has reached unprecedented levels. The consequences of unchecked fake news range from public health crises—as seen during the COVID-19 pandemic—to political manipulation, financial fraud, and incitement of violence.

This project presents a comprehensive machine learning–based system for the automated detection of fake news articles. Leveraging natural language processing (NLP) techniques and supervised classification algorithms, the system analyzes textual content to distinguish between genuine and fabricated news with high accuracy and efficiency.

The proposed solution employs a rigorous multi-model evaluation framework, systematically comparing four classification approaches: **Logistic Regression**, **Naive Bayes**, **Random Forest**, and **Long Short-Term Memory (LSTM)** neural networks. Feature extraction is performed using two complementary techniques: TF-IDF vectorization for classical machine learning models, and pre-trained GloVe word embeddings for the deep learning model. The data preprocessing pipeline includes tokenization, stopword removal, lemmatization, and noise filtering to ensure clean and consistent input representations.

Experimental results on two widely recognized benchmark datasets—the **LIAR dataset** (Wang, 2017) and the **Kaggle Fake News dataset**—demonstrate the effectiveness of the proposed approach. Among all models evaluated, the LSTM-based model achieves the highest accuracy of **95.4%**, with a precision of 94.8%, recall of 96.0%, and an F1-score of 95.1%. The Random Forest model provides an excellent trade-off between performance and computational efficiency at 93.8% accuracy, while Logistic Regression achieves a competitive 92.6% accuracy with minimal resource requirements.

The system is further deployed as a lightweight Flask-based web application, enabling real-time news verification for end users.

**Keywords:** Fake News Detection, Natural Language Processing, Machine Learning, LSTM, TF-IDF, Logistic Regression, Random Forest, Naive Bayes, Misinformation, Text Classification, Deep Learning, GloVe Embeddings, Flask Web Application.

## Chapter 1: Introduction

Fake news refers to deliberately fabricated information presented and disseminated in the guise of legitimate news content. It encompasses a wide spectrum of intentional misinformation, including hoaxes, fabricated political statements, sensationalized half-truths, and propaganda designed to mislead audiences for political, financial, or ideological purposes.

The explosive growth of social media platforms such as Facebook, Twitter/X, WhatsApp, and YouTube, combined with the decline of traditional journalistic gatekeeping, has created an information ecosystem in which any individual or organization can publish and disseminate content with global reach within seconds. Recommendation algorithms optimized for engagement rather than accuracy further amplify emotionally charged or sensational content, irrespective of its veracity.

The societal consequences of unchecked fake news are severe and well-documented:
- During the **2016 United States Presidential Election**, fabricated news stories were widely shared on social media, potentially influencing voter behavior.
- The **COVID-19 pandemic** demonstrated how medical misinformation could directly result in preventable deaths.
- In developing nations, fake news has been linked to communal violence, lynchings, and electoral manipulation.

Machine learning and natural language processing offer a promising path forward by training statistical models on large labeled datasets to automatically identify patterns that distinguish credible news from fabricated content.

## Chapter 2: Literature Review

### 2.1 Knowledge-Based Approaches
Vlachos and Riedel (2014) were among the first to formally define fact-checking as a computational problem. Popat et al. (2017) proposed CredEye, a system that assesses claim credibility by analyzing agreement signals across multiple web sources.

### 2.2 Linguistic and Stylometric Approaches
Perez-Rosas et al. (2018) demonstrated that fake news exhibits distinctive stylistic patterns — higher rates of informal language, stronger emotional language, and more grammatical errors (~74% accuracy with SVM). Horne and Adali (2017) found fake news uses fewer technical terms and more emotionally charged vocabulary.

### 2.3 Social Context and Network-Based Approaches
Castillo et al. (2011) pioneered credibility assessment of Twitter content. Ruchansky et al. (2017) proposed the CSI model combining text, user reactions, and user credibility. Shu et al. (2019) developed FakeDetector using multi-modal detection.

### 2.4 Machine Learning Approaches
Wang (2017) released the LIAR dataset and showed LSTM models outperform traditional classifiers. Ahmed et al. (2018) achieved 90–94% accuracy with LinearSVM and Logistic Regression on the Kaggle dataset. Granik and Mesyura (2017) applied Naive Bayes to Facebook fake news detection.

### 2.5 Deep Learning Approaches
Volkova et al. (2017) demonstrated bidirectional LSTM models achieve strong performance on rumor detection. CNNs (Kim, 2014) have also been applied to headline-level classification.

### 2.6 Transformer-Based Approaches
BERT (Devlin et al., 2019) set new benchmarks. Kula et al. (2020) fine-tuned BERT for fake news detection achieving 97–99% accuracy, but at significant computational cost.

### 2.7 Summary
Key findings: (1) classical ML with TF-IDF achieves 88–94% accuracy; (2) LSTM outperforms classical ML by capturing sequential dependencies; (3) transformers achieve near-human performance at high cost; (4) linguistic features are strong discriminators.

## Chapter 3: Problem Statement

### 3.1 Formal Problem Definition

The fake news detection task is defined as a **binary text classification** problem. Given a news article's textual content, the system classifies it as **REAL** (label 1) or **FAKE** (label 0).

Formally: Let $D = \{(x_1, y_1), (x_2, y_2), \ldots, (x_n, y_n)\}$ be a labeled dataset where:
- $x_i \in X$ is the textual feature vector
- $y_i \in \{0, 1\}$ is the binary label (0 = Fake, 1 = Real)

Goal: Learn classifier $f: X \rightarrow Y$ minimizing: $L(f) = \frac{1}{m} \sum_i \ell(f(x_i), y_i)$

### 3.2 Challenges
- **Class Imbalance**: Certain categories more prevalent than others
- **Stylistic Sophistication**: Modern fake news mimics legitimate news structure
- **Domain & Temporal Shift**: Models may not generalize across time periods
- **Label Noise**: Binary real/fake oversimplifies the credibility spectrum
- **Real-Time Requirements**: Low-latency processing needed for deployment

### 3.3 Success Criteria
- Overall accuracy ≥ 90%
- F1-score ≥ 89%
- Recall ≥ 90%
- Inference latency < 500ms per article

## Chapter 6: Methodology

### 6.1 Data Preprocessing Pipeline

The preprocessing pipeline consists of six sequential stages:

| Step | Operation | Description |
|------|-----------|-------------|
| 1 | Lowercasing | Convert all text to lowercase |
| 2 | Noise Removal | Remove URLs, HTML tags, special characters, digits |
| 3 | Tokenization | Split text into individual tokens using NLTK |
| 4 | Stopword Removal | Filter common English stopwords (179 words) |
| 5 | Stemming/Lemmatization | Reduce words to base form |
| 6 | Token Rejoining | Rejoin cleaned tokens into a single string |

### 6.2 Feature Extraction

**TF-IDF Vectorization** (for classical ML):
- `max_features=5000`, `ngram_range=(1,3)`
- Fitted on training corpus only to prevent data leakage

**GloVe Word Embeddings** (for LSTM):
- Pre-trained 100-dimensional vectors
- Vocabulary size: 50,000 words
- Sequence length: 500 tokens

### 6.3 Machine Learning Models

| Model | Type | Key Hyperparameters |
|-------|------|--------------------|
| Logistic Regression | Classical ML | C=1.0, max_iter=1000, solver='lbfgs' |
| Naive Bayes | Classical ML | MultinomialNB, alpha=0.1 |
| Random Forest | Ensemble ML | n_estimators=200, max_features='sqrt' |
| LSTM | Deep Learning | 2 LSTM layers (128, 64), Dropout=0.3, Adam lr=0.001 |

*Table 6: ML Model Hyperparameters*

### 6.4 LSTM Architecture

| Layer | Type | Parameters | Output Shape |
|-------|------|------------|-------------|
| 1 | Embedding | vocab=50000, dim=100, GloVe init | (None, 500, 100) |
| 2 | LSTM | units=128, return_sequences=True | (None, 500, 128) |
| 3 | Dropout | rate=0.3 | (None, 500, 128) |
| 4 | LSTM | units=64, return_sequences=False | (None, 64) |
| 5 | Dropout | rate=0.3 | (None, 64) |
| 6 | Dense | units=64, activation=ReLU | (None, 64) |
| 7 | Dense (Output) | units=1, activation=Sigmoid | (None, 1) |

*Table 7: LSTM Architecture Summary*

### 6.5 Preprocessing Code Implementation

````python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

nltk.download('stopwords', quiet=True)
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    """Complete text preprocessing pipeline."""
    if not isinstance(text, str):
        return ''
    # Step 1: Remove non-alphabetic characters
    text = re.sub('[^a-zA-Z]', ' ', text)
    # Step 2: Lowercasing
    text = text.lower()
    # Step 3: Tokenization
    tokens = text.split()
    # Step 4: Stopword removal + Step 5: Stemming
    tokens = [ps.stem(word) for word in tokens if word not in stop_words]
    # Step 6: Rejoin
    return ' '.join(tokens)

# Demonstrate preprocessing
sample_text = "WASHINGTON (Reuters) - The head of a conservative Republican faction in the U.S. Congress called himself a fiscal conservative."
print("Original Text:")
print(sample_text)
print("\nAfter Preprocessing:")
print(clean_text(sample_text))
````

### 7.3 Model Training Implementation

````python
import pandas as pd
import re
import pickle
import numpy as np
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')

ps = PorterStemmer()
stop_words = set(stopwords.words('english'))

# Load datasets
true = pd.read_csv('dataset/True.csv')
fake = pd.read_csv('dataset/Fake.csv')

# Assign labels
true['label'] = 1  # Real
fake['label'] = 0  # Fake

# Combine first 5000 of each
frames = [true.loc[:5000][:], fake.loc[:5000][:]]
df = pd.concat(frames)
df = df.dropna()
df2 = df.copy()
df2.reset_index(inplace=True)

print(f"Combined dataset shape: {df2.shape}")
print(f"Label distribution:\n{df2['label'].value_counts()}")
````

````python
# Text preprocessing
print("Preprocessing text data...")
corpus = []
for i in range(len(df2)):
    review = re.sub('[^a-zA-Z]', ' ', df2['text'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stop_words]
    review = ' '.join(review)
    corpus.append(review)

print(f"Corpus size: {len(corpus)} documents")
print(f"\nSample preprocessed text (first 200 chars):")
print(corpus[0][:200])
````

````python
# TF-IDF Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_v = TfidfVectorizer(max_features=5000, ngram_range=(1, 3))
X = tfidf_v.fit_transform(corpus).toarray()
y = df2['label']

print(f"TF-IDF Feature Matrix shape: {X.shape}")
print(f"Number of features: {X.shape[1]}")

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print(f"\nTraining set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")
````

````python
# Train PassiveAggressiveClassifier
from sklearn.linear_model import PassiveAggressiveClassifier

classifier = PassiveAggressiveClassifier(max_iter=1000)
classifier.fit(X_train, y_train)

pred = classifier.predict(X_test)
accuracy = metrics.accuracy_score(y_test, pred)

print(f"PassiveAggressive Classifier Results:")
print(f"Accuracy: {accuracy:.4f}")
print(f"\nClassification Report:")
print(metrics.classification_report(y_test, pred, target_names=['FAKE', 'REAL']))
````

````python
# Confusion Matrix Visualization
import matplotlib.pyplot as plt
import itertools

cm = metrics.confusion_matrix(y_test, pred)

fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
ax.set_title('Confusion Matrix — PassiveAggressive Classifier', fontsize=14, fontweight='bold')
plt.colorbar(im)
classes = ['FAKE', 'REAL']
tick_marks = np.arange(len(classes))
ax.set_xticks(tick_marks)
ax.set_xticklabels(classes)
ax.set_yticks(tick_marks)
ax.set_yticklabels(classes)

thresh = cm.max() / 2.
for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    ax.text(j, i, cm[i, j], horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black", fontsize=16)

ax.set_ylabel('True Label', fontsize=12)
ax.set_xlabel('Predicted Label', fontsize=12)
plt.tight_layout()
plt.savefig('report_images/confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
print(f"\nConfusion Matrix:\n{cm}")
````

### 8.6 Testing with User Inputs — Fake vs Real News

````python
import pickle
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

# Load the trained model and vectorizer
model = pickle.load(open('Fake-News-Detection-main/model2.pkl', 'rb'))
tfidfvect = pickle.load(open('Fake-News-Detection-main/tfidfvect2.pkl', 'rb'))

def predict_news(text):
    """Predict whether a news article is FAKE or REAL."""
    review = re.sub('[^a-zA-Z]', ' ', text)
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    review_vect = tfidfvect.transform([review]).toarray()
    prediction = 'FAKE' if model.predict(review_vect) == 0 else 'REAL'
    return prediction

print("Model loaded successfully!")
print(f"Model type: {type(model).__name__}")
print(f"Vectorizer vocabulary size: {len(tfidfvect.vocabulary_)}")
````

### A.1 Text Preprocessing Function

````python
# A.1 Text Preprocessing Function
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    """Complete text preprocessing pipeline for fake news detection."""
    if not isinstance(text, str): return ''
    text = text.lower()
    text = re.sub(r'http\S+|www\.\S+', '', text)   # Remove URLs
    text = re.sub(r'<[^>]+>', '', text)              # Remove HTML tags
    text = re.sub(r'[^a-z\s]', '', text)             # Keep only letters
    tokens = nltk.word_tokenize(text)                 # Tokenize
    tokens = [t for t in tokens if t not in stop_words]  # Remove stopwords
    tokens = [lemmatizer.lemmatize(t) for t in tokens]   # Lemmatize
    return ' '.join(tokens)

print("Preprocessing function defined.")
print(f"Stopwords count: {len(stop_words)}")
````

### A.2 LSTM Model Construction

````python
# A.2 LSTM Model Construction (Architecture Reference)
lstm_code = '''
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Embedding, LSTM,
                                     Dense, Dropout)

def build_lstm(embedding_matrix, vocab_size=50000,
               max_len=500, emb_dim=100):
    model = Sequential([
        Embedding(vocab_size, emb_dim,
                  weights=[embedding_matrix],
                  input_length=max_len,
                  trainable=False),
        LSTM(128, return_sequences=True),
        Dropout(0.3),
        LSTM(64),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model
'''
print("LSTM Model Architecture Code:")
print(lstm_code)
````
